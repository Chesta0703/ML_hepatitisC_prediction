```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns 
```

```python
df = pd.read_csv('./HepatitisCdata.csv')
df.head()
```

### Modifying data

```python
#Dropping Id Column
df.drop(['Unnamed: 0'], axis = 1, inplace=True)
```

```python
df.Category.value_counts()
```

```python
# df['Category'] = df['Category'].apply( lambda x: 1 if x in ['3=Cirrhosis','1=Hepatitis','2=Fibrosis'] else 0)
def caty(x):
    if x == '1=Hepatitis': 
        return 1
    elif x == '2=Fibrosis':
        return 2
    elif x == '3=Cirrhosis':
        return 3
    else:
        return 0
df['Category'] = df['Category'].apply( lambda x : caty(x))
```

```python
df.Category.value_counts()
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.median())
```

```python
df.Category.value_counts()
```

```python
df.isnull().sum()
```

```python
trainee_set = df.drop(['Category'], axis = 1)
```

```python
# df['Sex'] = df['Sex'].apply(lambda x: 1 if x == 'm' else 0)
# df.Sex.value_counts()

nominal_df = df[['Sex']]
nominal_df = pd.get_dummies(data=nominal_df)
nominal_df.head()
```

```python
df.drop(['Sex'], axis = 1, inplace = True)
df = df.join(nominal_df)
df.head()
```

```python
# checks if there is any relation with other variables
df.corr()
```

```python
#Correlation Heatmap
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(df.corr(),annot=True,cmap ="Blues_r")
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
x = df.loc[:, df.columns != 'Category'].values
# x= StandardScaler().fit_transform(x) # normalizing the features
```

```python
normalised_df = pd.DataFrame(x)
normalised_df.head()
```

```python
X = normalised_df      # data
Y= df['Category']      # target
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y,test_size = 0.3,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix

def hyper_tuning(name, model, model_params, X_train, y_train, X_test, y_test, cv):   # returns best model parameters
    grid_cv = GridSearchCV(model, [model_params], n_jobs=-1, cv=cv)
    grid_cv_model = grid_cv.fit(X_train, y_train)
    return grid_cv_model
```

### Model Selection

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier

model_params = {
    'knn' : {
        'model' : KNeighborsClassifier(),
        'params' : {
            "n_neighbors" : [5,15,25,30,35,40,100],
            "weights" : ["uniform" , "distance"],
        },
        'cv': 4
    },
    'logistic_regression' : {
        'model' : LogisticRegression(solver = 'liblinear', multi_class ='auto', random_state=42),
        'params' : {
            "penalty": ["l1", "l2"],
            "C": np.logspace(-2,4,100)
        },
        'cv' : 4
    },
    'random_forest_classifier' : {
        'model' :  RandomForestClassifier(n_estimators=100, random_state=42),
        'params' : {
            'max_features':[2, 3, 5, 7, 8],
        },
        'cv' : 5
    },
    
    'decision_tree_classifier' : {
        'model' : DecisionTreeClassifier(random_state=42),
        'params' : {
            'max_depth': [1, 2, 3, 4, 5, 8],
        },
        'cv': 4 
    },
    
    'extra_tree_classifier' : {
        'model' : ExtraTreesClassifier(random_state=42),
        'params' : {
            'n_estimators': [10, 50, 100],
            'max_depth': [1, 2, 3, 4, 5, 8],
        },
        'cv' : 5
    },
    
    'adaboost_classifier' : {
        'model' : AdaBoostClassifier(DecisionTreeClassifier(max_depth=8),random_state=42),
        'params' : {
            'n_estimators': [10, 50, 100],
            'learning_rate' : [0.25, 0.50]
        },
        'cv': 4
    }
}
```

```python
results = []

for model_name, mp in model_params.items():
    final_model = hyper_tuning(model_name, mp['model'], mp['params'], X_train, y_train, X_test, y_test, mp['cv'])
    results.append({
        'model_name' : model_name,
        'best_score' : final_model.best_score_,
        'best_params' : final_model.best_params_
    })

```

```python
final_matrix= pd.DataFrame(results, columns = ['model_name', 'best_score', 'best_params'])
final_matrix
```

#### From above search, we will use RandomForestClassifier considering both overfitting and underfitting cases

```python
final_model = GridSearchCV(RandomForestClassifier(n_estimators=100, random_state=42), {'max_depth': [1, 2, 3, 4, 5, 8]}, n_jobs=-1, cv=5)
final_model.fit(X_train, y_train)

def predict_disease(Age,ALB,ALP,ALT,AST,BIL,CHE,CHOL,CREA,GGT,PROT,Sex):
    
    x = np.zeros(shape= (1,13))
    
    if Sex==1:
        Sex_m = 1
        Sex_f = 0
    else:
        Sex_m = 0
        Sex_f = 1
        
    x[0] = [Age,ALB,ALP,ALT,AST,BIL,CHE,CHOL,CREA,GGT,PROT,Sex_f, Sex_m]
#     x = StandardScaler().fit_transform(x)
    
    return final_model.predict(x)[0]
```
